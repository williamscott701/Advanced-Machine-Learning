```python
def make_meshgrid(x, y, h=.02):
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))
    return xx, yy

def confusion_matri(y_truth, y_predicted):

    y_truth = np.array(y_truth)

    shape = max(np.unique(y_truth).shape[0], np.unique(y_predicted).shape[0]) +1
    
    m = [[0] * shape for i in range(shape)]
    
    for exp, pred in zip(y_truth, y_predicted):
        m[int(pred)][int(exp)] += 1
    
    n = shape-1
    for i in range(0,n):
        m[i][n] = sum(m[i])
    for j in range(n):
        t = 0
        for i in range(n):
            t = t + m[i][j]

        m[n][j] = t

    print("\nConfusion Matrix")
    for i in m:
        print(i)
    print("")
    return m
```

```python
def poly_kernel_2(x, y):
    c = 2
    d = 3
    t = np.dot(x, y.T) + c
#     print("shapes:", x.shape)
#     print(y.shape)
#     print(t.shape)
    return t**d
```

```python
def poly_predict_2(x, svm, support_vectors):
    
    t = poly_kernel_2(support_vectors, x)
    
    dual_coeff = svm.dual_coef_
    intercept = svm.intercept_

    mul = np.dot(dual_coeff, t)
    
    y_predicted = mul+intercept

    y_predicted[y_predicted>=0] = 1
    y_predicted[y_predicted<0] = 0

    return y_predicted
```

```python
data_n = temp_data[0]

x = np.array(data_n[0])
y = np.array(data_n[1])

svm = SVC(kernel=poly_kernel_2)
svm.fit(x, y)

support_indices = svm.support_

support_vectors = []
for index in support_indices:
    support_vectors.append(x[index])

support_vectors = np.array(support_vectors)

y_predicted = poly_predict_2(x, svm, support_vectors)

X0, X1 = x[:, 0], x[:, 1]
xx, yy = make_meshgrid(X0, X1)
Z = poly_predict_2(np.c_[xx.ravel(), yy.ravel()], svm, support_vectors)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.winter)
plt.scatter(X0, X1, c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.show()

print("Score:", svm.score(x, y))
print("F1 Score:",f1_score(y, y_predicted.T, average='weighted'))
print("Accuracy:", accuracy_score(y, y_predicted.T, normalize=False))
```

```python
def linear_predict(x, w, b):
    y_predicted = np.dot(x, w.T) + b
    
    y_predicted[y_predicted>=0] = 1
    y_predicted[y_predicted<0] = 0
    return y_predicted

def linear_predict_multiclass_convert(y):
    y_predic = np.array(y).T[0]

    y_predicted = []
#     print(y_predic)
    
    for y_data in y_predic:
        y_predicted.append(np.argmax(y_data, axis = 0))
        
    return np.array(y_predicted)

def one_vs_rest_get_svm(data_in):

    svm_s = []

    for i in np.unique(data_in[1]):
        print("Processing Class y =", i)
        data_x = np.array(data_in[0])
        data_y = np.array(data_in[1])

        data_y_ = copy.deepcopy(data_y)

        data_y_[data_y_== i] = 4
        data_y_[data_y_!= 4] = 0
        data_y_[data_y_== 4] = 1

        model = SVC(kernel='linear')
        model.fit(data_x, data_y_)
        svm_s.append(model)

    return svm_s

def linear_predict_ovr(test_x, svm_s):
    y_predic = []

    for clf in svm_s:
        b = linear_predict(test_x, clf.coef_, clf.intercept_)
        y_predic.append(b)

    y_multiclass = linear_predict_multiclass_convert(y_predic)
    
    return y_multiclass, y_predic

def one_vs_rest(X, Y):

    svm_s = one_vs_rest_get_svm([X, Y])

    y_predicted, y_singles = linear_predict_ovr(X, svm_s)

    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')

    X0, X1 = X[:, 0], X[:, 1]
    xx, yy = make_meshgrid(X0, X1)

    Z = linear_predict_ovr(np.c_[xx.ravel(), yy.ravel()], svm_s)[0]
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z)

    plt.show()
    
data = temp_data[2]

X = np.array(data[0])
Y = np.array(data[1])
one_vs_rest(X, Y)
```

```python
def general_svm(x, y, kernel):
    svm = SVC(kernel=kernel)
    svm.fit(x, y)
    score = svm.score(x, y)
    w = svm.coef_
    b = svm.intercept_
    return w, b, score, svm
```

```python
def linear_soft_margin(data, X, Y, x_train, x_test, y_train, y_test):
    w, b, score, model = general_svm(x_train, y_train, 'linear')

    y_predicted, y_decision_b = linear_predict(x_test, w, b)

    print("F1 Score:",f1_score(y_test, y_predicted, average='weighted'))
    print("Train Score:", score)
    print("Test Accuracy:", accuracy_score(y_test, y_predicted, normalize=False))
    print(y_test[:10], list(y_predicted)[:10])
    confusion_matri(y_test, y_predicted)

    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')

    plt.axis('tight')
    x_min = -3
    x_max = 3
    y_min = -3
    y_max = 3

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    
    z_predicted, Z = linear_predict(np.c_[XX.ravel(), YY.ravel()], w, b)

    Z = Z.reshape(XX.shape)
    plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    plt.contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-b, 0, b])

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    plt.xticks(())
    plt.yticks(())
    plt.show()
    
    _, y_score = linear_predict(x_test, w, b)
    plot_roc(y_test, y_score, 'Test')
    
    _, y_score = linear_predict(x_train, w, b)
    plot_roc(y_train, y_score, 'Train')

def plot_roc(y_test, y_score, name):
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, color='darkorange', lw=2)

    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',label='ROC '+str(name)+' (area = %0.2f)' % roc_auc)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

# PART 3 - LINEAR ONE vs ONE

```python
def linear_predict(x, w, b):
    y_predicted_decision = np.dot(x,w.T) + b
    
    y_predicted = copy.deepcopy(y_predicted_decision)
    
    y_predicted[y_predicted>=0] = 1
    y_predicted[y_predicted<0] = 0
    return y_predicted, y_predicted_decision


def one_vs_one_get_svm(data):

    svm_s = []
    combinations = []
    for i in np.unique(data[1]):
        for j in np.unique(data[1]):
            if i < j:
                combinations.append([i,j])
                print("y =", i, j)

                data_x = []
                data_y_ = []

                for x, y in zip(data[0], data[1]):
                    if y == i or y == j:
                        data_x.append(x)
                        data_y_.append(y)

                data_y = np.array(data_y_)

                data_y[data_y== i] = 4
                data_y[data_y!= 4] = 1
                data_y[data_y== 4] = 0

                svm = SVC(kernel="linear")
                svm.fit(data_x, data_y)
                svm_s.append(svm)
        
    return svm_s, combinations


def linear_predict_ovo(test_x, svm_s, combinations):
    y_predic = []
    y_dec = []

    for clf in svm_s:
        b, c = linear_predict(test_x, clf.coef_, clf.intercept_)
        y_predic.append(b)
        y_dec.append(c)

    y_multiclass_pred, y_multiclass_dec = linear_predict_multiclass_convert_ovo(y_predic, y_dec, combinations)
    
    return y_multiclass_pred, y_multiclass_dec


def linear_predict_multiclass_convert_ovo(y, y_dec, combinations):
    y = np.array(y).T[0]

    y_dec = np.array(y_dec).T[0]
    
    y_predicted = []
    y_d = []
    
    for p in y:
        val = [0,0,0]
        for i in range(0, len(p)):
            if p[i] <= 0:
                val[combinations[i][0]]= val[combinations[i][0]]+1
            else:
                val[combinations[i][1]]= val[combinations[i][1]]+1
        y_predicted.append(val.index(max(val)))

    if len(combinations)>1:
        for y, dec in zip(y_predicted, y_dec):
            y_d.append(dec[y])
    else:
        y_d = y_dec
    
    
    return np.array(y_predicted), np.array(y_d)


def one_vs_one(X, Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

    svm_s, combinations = one_vs_one_get_svm([x_train, y_train])

    y_predicted, y_dec = linear_predict_ovo(x_test, svm_s, combinations)

    print("\nF1 Score:",f1_score(y_test, y_predicted, average='weighted'))
    print("Accuracy:", accuracy_score(y_test, y_predicted, normalize=False))
    print(y_test[:10], list(y_predicted)[:10])
    confusion_matri(y_test, y_predicted)

    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')

    X0, X1 = X[:, 0], X[:, 1]
    xx, yy = make_meshgrid(X0, X1)

    Z, _ = linear_predict_ovo(np.c_[xx.ravel(), yy.ravel()], svm_s, combinations)
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z)

    plt.show()
    
    y_test_ = copy.deepcopy(y_test)
    y_test_[y_test_>1]=1
    
    plot_roc(y_test_, y_dec, 'Test')
    
```

# PART 4 - RBF ONE VS REST

```python

def one_vs_rest_get_svm(data_in):

    svm_s = []
    
    support_vectors_all = []

    for i in np.unique(data_in[1]):
        print("Processing Class y =", i)
        data_x = np.array(data_in[0])
        data_y = np.array(data_in[1])

        data_y_ = copy.deepcopy(data_y)

        data_y_[data_y_== i] = 4
        data_y_[data_y_!= 4] = 0
        data_y_[data_y_== 4] = 1

        model = SVC(kernel='rbf')
        model.fit(data_x, data_y_)
        svm_s.append(model)
        
        support_indices = model.support_

        support_vectors = []
        for index in support_indices:
            support_vectors.append(data_x[index])

        support_vectors = np.array(support_vectors)
        
        support_vectors_all.append(support_vectors)
        
    return svm_s, support_vectors_all

def rbf_kernel_ws(X,Y):
    print(X.shape[0],Y.shape[0])
    K = np.zeros((X.shape[0],Y.shape[0]))
    for i,x in enumerate(X):
        for j,y in enumerate(Y):
            K[i,j] = np.linalg.norm(x-y)**2
            if i%10==0 and j %50000==0:
                print(i,j)
    sigma = 1
    return np.exp(-(1.0/2.0*(sigma**2))*K)

def rbf_predict(x, svm, support_vectors):
    
    t = rbf_kernel(support_vectors, x)
    
    dual_coeff = svm.dual_coef_
    intercept = svm.intercept_

    mul = np.dot(dual_coeff, t)
    
    y_predicted = mul+intercept

    y_predicted[y_predicted>=0] = 1
    y_predicted[y_predicted<0] = 0

    return y_predicted

def rbf_predict_multiclass_convert(y):

    y_predic = np.array(y).T

    y_predicted = []
    
    for y_data in y_predic:
        y_predicted.append(np.argmax(y_data[0], axis = 0))
        
    return np.array(y_predicted)

def rbf_predict_ovr(test_x, svm_s, support_vectors):
    y_predic = []
    
    for clf, s_v in zip(svm_s, support_vectors):
        b = rbf_predict(test_x, clf, s_v)
        y_predic.append(b)

    y_multiclass = rbf_predict_multiclass_convert(y_predic)
    
    return y_multiclass

def one_vs_rest_rbf(X, Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

    svm_s, support_vectors = one_vs_rest_get_svm([x_train, y_train])
    
    y_predicted = rbf_predict_ovr(x_test, svm_s, support_vectors)

    print("\nF1 Score:",f1_score(y_test, y_predicted, average='weighted'))
    print("Accuracy:", accuracy_score(y_test, y_predicted, normalize=False))
    print(y_test[:10], list(y_predicted)[:10])
    confusion_matri(y_test, y_predicted)

    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')

    X0, X1 = X[:, 0], X[:, 1]
    xx, yy = make_meshgrid(X0, X1)

    print(np.c_[xx.ravel(), yy.ravel()].shape)
    
    Z = rbf_predict_ovr(np.c_[xx.ravel(), yy.ravel()], svm_s, support_vectors)
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z)

    plt.show()
    
    y_train[y_train>1]=1
    plot_roc(y_train, svm_s[0].decision_function(x_train), 'Test')
```

# PART 4 - RBF ONE VS ONE

```python

def ovo_get_svm_rbf(data_in):

    svm_s = []
    combinations = []
    
    support_vectors_all = []
    
    for i in np.unique(data[1]):
        for j in np.unique(data[1]):
            if i < j:
                combinations.append([i,j])
                print("y =", i, j)

                data_x = []
                data_y_ = []

                for x, y in zip(data[0], data[1]):
                    if y == i or y == j:
                        data_x.append(x)
                        data_y_.append(y)

                data_y = np.array(data_y_)

                data_y[data_y== i] = 4
                data_y[data_y!= 4] = 1
                data_y[data_y== 4] = 0

                model = SVC(kernel="rbf")
                model.fit(data_x, data_y)
                svm_s.append(model)

                support_indices = model.support_

                support_vectors = []
                for index in support_indices:
                    support_vectors.append(data_x[index])

                support_vectors = np.array(support_vectors)

                support_vectors_all.append(support_vectors)
        
    return svm_s, support_vectors_all, combinations

def rbf_kernel_ws(X,Y):
    print(X.shape[0],Y.shape[0])
    K = np.zeros((X.shape[0],Y.shape[0]))
    for i,x in enumerate(X):
        for j,y in enumerate(Y):
            K[i,j] = np.linalg.norm(x-y)**2
            if i%10==0 and j %50000==0:
                print(i,j)
    sigma = 1
    return np.exp(-(1.0/2.0*(sigma**2))*K)

def rbf_predict(x, svm, support_vectors):
    
    t = rbf_kernel(support_vectors, x)
    
    dual_coeff = svm.dual_coef_
    intercept = svm.intercept_

    mul = np.dot(dual_coeff, t)
    
    y_predicted = mul+intercept

    y_predicted[y_predicted>=0] = 1
    y_predicted[y_predicted<0] = 0

    return y_predicted

def rbf_predict_multiclass_convert_ovo(y, combinations):
    y = np.array(y).T

    y_predicted = []
    
    for p in y:
        val = [0,0,0]
        for i in range(0, len(p)):
            if p[i] <= 0:
                val[combinations[i][0]]= val[combinations[i][0]]+1
            else:
                val[combinations[i][1]]= val[combinations[i][1]]+1
        y_predicted.append(val.index(max(val)))
        
    return np.array(y_predicted)


def rbf_predict_ovo(test_x, svm_s, support_vectors, combinations):
    y_predic = []
    
    for clf, s_v in zip(svm_s, support_vectors):
        b = rbf_predict(test_x, clf, s_v)
        y_predic.append(b)
        

    y_multiclass = rbf_predict_multiclass_convert_ovo(y_predic, combinations)
    print(y_multiclass)
    
    return y_multiclass

def one_vs_one_rbf(X, Y):
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

    svm_s, support_vectors, combinations = ovo_get_svm_rbf([x_train, y_train])
    
    y_predicted = rbf_predict_ovo(x_test, svm_s, support_vectors, combinations)
    
    print(y_predicted)

    print("\nF1 Score:",f1_score(y_test, y_predicted, average='weighted'))
    print("Accuracy:", accuracy_score(y_test, y_predicted, normalize=False))
    print(y_test[:10], list(y_predicted)[:10])
    confusion_matri(y_test, y_predicted)
    
    plt.scatter(X[:, 0], X[:, 1], c=Y, zorder=10, cmap=plt.cm.Paired, edgecolors='k')

    X0, X1 = X[:, 0], X[:, 1]
    xx, yy = make_meshgrid(X0, X1)

    print(np.c_[xx.ravel(), yy.ravel()].shape)
    
    Z = rbf_predict_ovo(np.c_[xx.ravel(), yy.ravel()], svm_s, support_vectors, combinations)
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z)

    plt.show()
    
    y_train[y_train>1]=1
    plot_roc(y_train, svm_s[0].decision_function(x_train), 'Test')
```
